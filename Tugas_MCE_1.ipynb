### Convert Waktu

Melakukan konversi terhadap rentang waktu yang diberikan. diberikan contoh dengann kasus

---
Misal :
* bulan Sekarang = Januari 2020
* bulan pensiun = Juli 2020
* bulan akhir pensiun = Desember 2020
---
Maka setoran dilakukan hanya sampai bulan Juni 2020, yaitu sebanyak 6 kali. Dan pembayaran selama masa pensiun adalah 6 kali terhitung sejak Juli 2020 sampai Desember 2020.




```
def bayar(bulan_awal,tahun_awal,bulan_akhir,tahun_akhir):
  bulan_awal = bulan_awal[:3].lower()
  bulan_akhir = bulan_akhir[:3].lower()

  x = {"jan" : 1, "feb" : 2, "mar" : 3, "apr" : 4,"mei" : 5,"jun" : 6 , "jul" : 7, "agu" : 8 , "sep" : 9, "okt" : 10 , "nov" : 11 , "des" : 12}

  ## case tahun sama

  if tahun_awal == tahun_akhir :
    if x[bulan_awal] >= x[bulan_akhir] :
      return -1
    else :
      return x[bulan_akhir]-x[bulan_awal]+1 
  
  ## case tahun beda
  
  elif tahun_awal < tahun_akhir :
    if x[bulan_awal] >= x[bulan_akhir] :
      return (12-x[bulan_awal]+1) + ((tahun_akhir-tahun_awal-1)*12) + x[bulan_akhir]
    else :
      return (12-x[bulan_awal]+1)+ ((tahun_akhir-tahun_awal-1)*12) + (x[bulan_akhir]-x[bulan_awal]+1)
  else :
    return -1 

```

### Interval Waktu (m dan n)

---
* **bulan_sekarang**  = Bulan pada saat dilakukan perhitungan (sekarang)
* **tahun_sekarang**  = Tahun pada saat dilakukan perhitungan (sekarang)
* **bulan_pensiun**   = Bulan pada saat dimulainya masa pensiun
* **tahun_pensiun**   = Bulan pada saat dimulainya masa pensiun
* **bulan_akhir**     = Bulan pada saat selesainya masa pensiun
* **tahun_akhir**     = Bulan pada saat selesainya masa pensiun
---
Sehingga dihasilkan :
* **m** = banyaknya pembayaran selama masa pensiun (dalam bulan) 
* **n** = banyaknya setoran yang harus dilakukan (dalam bulan)
---


```
#@title Input { run: "auto", output-height: 10, form-width: "700px", display-mode: "both" }
bulan_sekarang = "Januari" #@param ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"]
tahun_sekarang = 2020#@param {type:"slider", min:2020, max:2100, step:1}
bulan_pensiun = "Juni" #@param ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"]
tahun_pensiun = 2020 #@param {type:"slider", min:2020, max:2100, step:1}
bulan_akhir = "Desember" #@param ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"]
tahun_akhir = 2020 #@param {type:"slider", min:2020, max:2100, step:1}


if tahun_pensiun >= tahun_sekarang and tahun_pensiun <= tahun_akhir :
  m = int(bayar(bulan_pensiun,tahun_pensiun,bulan_akhir,tahun_akhir))
  n = int(bayar(bulan_sekarang,tahun_sekarang,bulan_pensiun,tahun_pensiun) - 1)
else :
  m = -1
  n = -1

step = n
steps = m
print("m =", m)
print("n = ",n)
```

Proses 



>



1. periksa apakah nilai **m** dan **n** memenuhi (lebih dari 0), jika **ya** : 

> 1. Cari nilai *Present Value Anuitas* dengan menggunakan rumus yang diberikan di atas
2. Cari nilai P yang didapat dari hasil perkalian **C** dengan **PA**
3. Didapatkan nilai **S** dari hasil pembagian **P** terhadap **n**

2. jika **tidak**, maka tidak bisa dilanjutkan




```
list_pay = []
n = step
turun = 1
while n > 0 :
  m = steps
  while m > 0 :
    if m > 0 and n > 0 :

      # Proses 1

      PA = (1-(pow((1+r),-1*(m)))) / r
      PA = round(PA,4)
      

      # Proses 2

      P = C*PA
      P = round(P,2)
      
      # Proses 3

      #S = P/n
      # Proses 3 Revisi

      S = P/(1+r)**(n)
      S = S/n
      S = round(S,2)

      # Mmenghitung total uang disetor

      total = S*n
      
      print(S)
      #print("n = ",n)
      #print(PA)

      list_pay.append([r*12*100,PA,S,n,total,P,C,m,C*m,C*m-total])
    else :
        print("Tahun Pensiun <= Tahun Sekarang")
    m -= turun
  n -= turun



print(list_pay)
```
